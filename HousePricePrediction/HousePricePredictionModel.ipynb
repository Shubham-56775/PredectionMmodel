```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('/content/house_prices.csv')
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['bathrooms'] = df['bathrooms'].fillna(df['bathrooms'].median())
df.isnull().sum()
```

```python
q1 = df['price'].quantile(0.25)
q3 = df['price'].quantile(0.75)
iqr = q3 - q1
upper = q3 + 3 * iqr
df = df[df['price']<=upper]
df.shape
```

```python
x = df.drop(columns = 'price')
y = df['price']
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
x_std = scaler.transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
cf = RandomForestRegressor(n_estimators=200, max_depth = 10,random_state = 2 )
```

```python
cf.fit(x_train,y_train)
```

```python
x_train_pred = cf.predict(x_train)
```

```python
x_train_r2 = r2_score(y_train,x_train_pred)
```

```python
print(x_train_r2)
```

```python
x_test_pred = cf.predict(x_test)
x_test_r2 = r2_score(y_test, x_test_pred)
```

```python
print(x_test_r2)
```
